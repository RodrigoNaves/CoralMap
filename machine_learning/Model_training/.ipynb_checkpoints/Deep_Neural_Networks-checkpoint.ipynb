```python
#design a time-series multi-target regression algorithm
#split data randomly
data_2006 = pd.read_csv('x_df_2006_data.csv')
data_2011 = pd.read_csv('x_df_2011_data.csv')
data_2016 = pd.read_csv('merged_2016_data.csv')
y_labels_2006 = pd.read_csv('y_for_2006_data2011.csv')
y_labels_2011 = pd.read_csv('y_for_2011_data2016.csv')
```

```python
data_2006 = data_2006.dropna(how='any', axis=1)
data_2011 = data_2011.dropna(how='any', axis=1)
data_2011.head()
```

```python
#shuffle data from 2006 and 2011 in order to split training and testing data
data_x = [data_2006, data_2011]
x = pd.concat(data_x)

data_y = [y_labels_2006, y_labels_2011]
y_a = pd.concat(data_y)
#shuffle
X, y = shuffle(x, y_a, random_state=13)
#separate the LGA names since it is not needed for running model
#note that the LGA codes have been shuffled with the data so they will still follow the order of the model data
X_a = X.iloc[:,2:]
lgas = X.iloc[:,:2]
```

```python
X = X_a.astype(np.float32)
y = y.astype(np.float32)
X,y = np.array(X), np.array(y) #convert to numpy array
#print(X.shape)

print(X.shape)
print(y.shape)
np.any(np.isnan(X))
```

```python




X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_test = X_test.reshape((len(X_test), 599,1))
X_train = X_train.reshape((len(X_train), 599,1))
```

```python
#LSTM model

LSTMmodel = keras.Sequential()
# Add an Embedding layer expecting input vocab of size 1000, and
# output embedding dimension of size 64.
LSTMmodel.add(layers.SimpleRNN(500, input_shape=(599,1), unroll=True, activation='linear'))
LSTMmodel.add(layers.LeakyReLU())
'''
LSTMmodel.add(layers.Embedding(input_dim=603, output_dim=300))

# Add a LSTM layer with 128 internal units.
LSTMmodel.add(layers.LSTM(128))
#LSTMmodel.add(layers.Embedding(input_dim=128, output_dim=1200))
LSTMmodel.add(layers.LSTM(1200))
'''

LSTMmodel.add(layers.Dense(550,activation='linear'))
LSTMmodel.add(layers.LeakyReLU())
LSTMmodel.add(layers.Dense(450,activation='linear'))
LSTMmodel.add(layers.LeakyReLU())
# Add a Dense layer with 10 units.
LSTMmodel.add(layers.Dense(266,activation='linear'))

LSTMmodel.summary()


```

```python
LSTMmodel.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
           loss= tf.losses.MeanSquaredError(),metrics=[tf.metrics.MeanAbsoluteError()])
```

```python
%load_ext tensorboard
from datetime import datetime
from packaging import version
import tensorboard
print(tensorboard.__version__)
logdir="logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)
history_LSTM = LSTMmodel.fit(X_train, y_train, validation_data=(X_test,y_test),epochs=250, batch_size=50,shuffle=False,callbacks=[tensorboard_callback])

```

```python
import matplotlib.pyplot as plt
# plot the training and testing accuracies under different epochs
plt.plot(history_LSTM.history['mean_absolute_error'], label='Train')
plt.plot(history_LSTM.history['val_mean_absolute_error'], label='Test')
plt.ylabel('mean_absolute_error')
plt.xlabel('Epoch')
plt.title('Training and testing error')
plt.legend()
plt.show()
plt
```

```python
train_score = r2_score(y_train,y_train_preds,multioutput='uniform_average')
test_score = r2_score(y_test,y_test_preds,multioutput='uniform_average')
print("train r2 is:",train_score,"test r2 is:",test_score)
```
