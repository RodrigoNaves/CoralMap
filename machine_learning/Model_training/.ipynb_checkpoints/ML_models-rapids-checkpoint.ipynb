```python
#import lazypredict
#from lazypredict.Supervised import LazyRegressor
import pandas as pd
import numpy as np
import bamboolib
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.dummy import DummyRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
data_2006_x = pd.read_csv('x_df_2006_data.csv')
data_2006_y = pd.read_csv('y_for_2006_data2011.csv')
data_2011_x = pd.read_csv('x_df_2011_data.csv')
data_2011_y = pd.read_csv('y_for_2011_data2016.csv')
```

```python
data_2006_x = data_2006_x.dropna(how='any', axis=1)
data_2006_x
```

```python
data_2011_x = data_2011_x.dropna(how='any', axis=1)
data_2011_x
```

```python
data_2006_y
```

## Dummy classifier for time series
## Preprocess data normalizing, scaling and run Lazy predict again

```python
#shuffle data from 2006 and 2011 in order to split training and testing data
data_x = [data_2006_x, data_2011_x]
x = pd.concat(data_x)

data_y = [data_2006_y, data_2011_y]
y_a = pd.concat(data_y)
#shuffle
X, y = shuffle(x, y_a, random_state=13)
#separate the LGA names since it is not needed for running model
#note that the LGA codes have been shuffled with the data so they will still follow the order of the model data
X_a = X.iloc[:,2:]
lgas = X.iloc[:,:2]
```

```python
X = X_a.astype(np.float32)
y = y.astype(np.float32)
np.any(np.isnan(X))
```

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

'''
#scale training data
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)
'''
X_scaled = X_train
```

```python
#baseline dummy model, estimating refression using the mean values
dummy_baseline_model = MultiOutputRegressor(DummyRegressor(strategy="mean")).fit(X_scaled,y_train)

dummy_predictions = dummy_baseline_model.predict(X_test)

dummy_score = r2_score(y_test,dummy_predictions,multioutput='uniform_average')
dummy_score
```

## Regression algorithms to test
#### Random Forest regression
#### Lasso Regression
#### Kneighbours regreesion
#### Gaussian Processes regression
#### XGBRegressor
#### Support Vector Regression


```python
#models, predictions = MultiOutputRegressor(LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)).fit(X_scaled, X_test, y_train, y_test)
#Random Forests Regression
random_forest_model = MultiOutputRegressor(RandomForestRegressor()).fit(X_scaled,y_train)
```

```python
random_forest_predictions = random_forest_model.predict(X_test)
random_forest_score = r2_score(y_test,random_forest_predictions,multioutput='uniform_average')



random_forest_train_predictions = random_forest_model.predict(X_train)
random_forest_train_score = r2_score(y_train,random_forest_train_predictions,multioutput='uniform_average')
print("train acuracy is:",random_forest_train_score,"test acuracy is:",random_forest_score)
```

```python
#K Neighbours regressor
KNN_reg_model = MultiOutputRegressor(KNeighborsRegressor()).fit(X_scaled,y_train)
```

```python
KNN_predictions = KNN_reg_model.predict(X_test)
KNN_score = r2_score(y_test,KNN_predictions,multioutput='uniform_average')



KNN_train_predictions = KNN_reg_model.predict(X_train)
knn_train_score = r2_score(y_train,KNN_train_predictions,multioutput='uniform_average')
print("train acuracy is:",knn_train_score,"test acuracy is:",KNN_score)
```

```python

GP_reg_model = MultiOutputRegressor(GaussianProcessRegressor()).fit(X_scaled,y_train)
```

```python
GP_predictions = GP_reg_model.predict(X_test)
GP_score = r2_score(y_test,GP_predictions,multioutput='uniform_average')


GP_train_predictions = GP_reg_model.predict(X_train)
GP_train_score = r2_score(y_train,GP_train_predictions,multioutput='uniform_average')
print("train acuracy is:",GP_train_score,"test acuracy is:",GP_score)
```

```python
xgb_model = MultiOutputRegressor(XGBRegressor()).fit(X_scaled,y_train)
```

```python
xgb_predictions = xgb_model.predict(X_test)
xgb_score = r2_score(y_test,xgb_predictions,multioutput='uniform_average')


xgb_train_predictions = xgb_model.predict(X_train)
xgb_train_score = r2_score(y_train,xgb_train_predictions,multioutput='uniform_average')
print("train acuracy is:",xgb_train_score,"test acuracy is:",xgb_score)
```

```python
svr_model = MultiOutputRegressor(SVR()).fit(X_scaled,y_train)
```

```python
svr_predictions = svr_model.predict(X_test)
svr_score = r2_score(y_test,svr_predictions,multioutput='uniform_average')


svr_train_predictions = svr_model.predict(X_train)
svr_train_score = r2_score(y_train,svr_train_predictions,multioutput='uniform_average')
print("train acuracy is:",svr_train_score,"test acuracy is:",svr_score)
```
