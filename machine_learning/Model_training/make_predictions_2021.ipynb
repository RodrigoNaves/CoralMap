```python
from joblib import dump, load
import pandas as pd
import numpy as np
import bamboolib
```

```python
#import best regression model to predict 2021
clf = load("random_forest_clf.joblib")
```

```python
#import traning data t-1 for labels in 2021 at t
data_2016_x = pd.read_csv('x_df_2016_data.csv')
#x_attributes = pd.read_csv('x_shape_mould.csv')

data_2016_x
```

```python
#separate the lga names from the data

#x_2016 = result1.iloc[:604,:]
X_a = data_2016_x.iloc[:,2:]
lgas = data_2016_x.iloc[:,:2]
X = X_a.dropna(how='any', axis=1)
X = X_a.astype(np.float32)
np.any(np.isnan(X))
```

```python
predictions_2021 = clf.predict(X)
```

```python
predictions_2021.shape
```

```python
#get the column names and add them to the predictions_2021 numpy array
data_y = pd.read_csv("y_2016.csv")
data_y = data_y.drop(columns=['Unnamed: 0'])
data_y =  data_y.iloc[:,1:]
data_y = data_y.loc[:0,:]
#data_y = pd.concat([data_y, predictions_2021], ignore_index=True, sort=False)
df_predictions = pd.DataFrame(data=predictions_2021, columns=[list(data_y)])
```

```python
df_predictions.columns = ["_".join([str(index) for index in multi_index]) for multi_index in df_predictions.columns.ravel()]
df_predictions
```

```python
lgas = lgas.drop(columns=['Unnamed: 0'])
```

```python
lga = lgas["lga"]
df_predictions = df_predictions.join(lga)
df_predictions = df_predictions[['lga'] + [x for x in df_predictions.columns if x not in ['lga']]]
df_predictions
```
