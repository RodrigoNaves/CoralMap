```python
data_2006_x = pd.read_csv('../Model_training/x_df_2006_data.csv')
data_2006_y = pd.read_csv('y_for_2006_data2011.csv')
data_2011_x = pd.read_csv('x_df_2011_data.csv')
data_2011_y = pd.read_csv('y_for_2011_data2016.csv')
data_2016_x = pd.read_csv('x_df_2016_data.csv')
```

## Dummy classifier for time series


```python
#join 2006 and 2011 data
data_x = [data_2006_x, data_2011_x]
x = pd.concat(data_x)
x = x.dropna(how={'any'},thresh = 1000, axis=1)
x
```

```python
#shuffle data from 2006 and 2011 in order to split training and testing data

data_y = [data_2006_y, data_2011_y]
y_a = pd.concat(data_y)
#shuffle
X, y = shuffle(x, y_a, random_state=13)
#separate the LGA names since it is not needed for running model
#note that the LGA codes have been shuffled with the data so they will still follow the order of the model data
X_a = X.iloc[:,2:]
lgas = X.iloc[:,:2]

y_a =  y.iloc[:,1:]
```

```python
#drop values that are not found in both census 2006 and 2011
X = X_a.dropna(how='any', axis=1)
X
```

```python
#ensure data type is float and check that there are no nan values
X = X_a.astype(np.float32)
y_a = y_a.astype(np.float32)
np.any(np.isnan(X))
```

```python
#increase Polynomial of data to increase the chances of fitting data
#this method is not used,since linear data was enough to find a fitting model
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(interaction_only=True)
X_poly = poly.fit_transform(X)

X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.2)
X_train_poly.shape
```

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(X, y_a, test_size=0.2)

#scale is not necessary since data is already into a normal distribution of values, and there are no outliers
'''
#scale training data
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)
'''
X_scaled = X_train
X_scaled.shape
```

```python
#baseline dummy model, estimating refression using the mean values
dummy_baseline_model = MultiOutputRegressor(DummyRegressor(strategy="mean")).fit(X_scaled,y_train)

dummy_predictions = dummy_baseline_model.predict(X_test)

dummy_score = r2_score(y_test,dummy_predictions,multioutput='uniform_average')
dummy_score
```

```python
#models, predictions = MultiOutputRegressor(LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)).fit(X_scaled, X_test, y_train, y_test)
#Random Forests Regression
random_forest_model = MultiOutputRegressor(RandomForestRegressor(n_estimators = 150,warm_start=True,n_jobs= -1)).fit(X_scaled,y_train)
```

```python
random_forest_predictions = random_forest_model.predict(X_test)
random_forest_score = r2_score(y_test,random_forest_predictions,multioutput='uniform_average')

random_forest_train_predictions = random_forest_model.predict(X_train)
random_forest_train_score = r2_score(y_train,random_forest_train_predictions,multioutput='uniform_average')
print("train r2 is:",random_forest_train_score,"test r2 is:",random_forest_score)
```

```python
#Random Forrest using Cross Validation
from sklearn.model_selection import cross_validate
from sklearn import metrics
scoring = ['r2','neg_mean_absolute_error','neg_mean_squared_error']
scores = cross_validate(random_forest_model, X, y_a, cv=5, scoring = scoring)
scores
```

```python
#Random Forrest using a pipeline that scales, transforms and runs models in CV
from sklearn.decomposition import PCA
pca = PCA(n_components=100)
x_pca = pca.fit_transform(X)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_pca, y_a, test_size=0.2)
random_forest_model_pca = MultiOutputRegressor(RandomForestRegressor(n_estimators = 150,n_jobs= -1)).fit(X_train_pca,y_train)




```

```python
from sklearn.metrics import mean_absolute_error

random_forest_train_predictions_pca = random_forest_model_pca.predict(X_train_pca)
random_forest_train_score_pca = r2_score(y_train_pca,random_forest_train_predictions_pca,multioutput='uniform_average')
random_forest_predictions_pca = random_forest_model_pca.predict(X_test_pca)
random_forest_score_pca = r2_score(y_test_pca,random_forest_predictions_pca,multioutput='uniform_average')

random_forest_train_scoremae_pca = mean_absolute_error(y_train_pca,random_forest_train_predictions_pca,multioutput='uniform_average')
random_forest_scoremae_pca = mean_absolute_error(y_test_pca,random_forest_predictions_pca,multioutput='uniform_average')

print("train r2 is:",random_forest_train_score_pca,"test r2 is:",random_forest_score_pca)
print("train mean absolute error is:",random_forest_train_scoremae_pca,"test mean absolute error is:",random_forest_scoremae_pca)
```
